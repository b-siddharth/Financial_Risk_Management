```python
from scipy.stats import norm
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.5, 1)})

median=house_data.area.median()
mean=house_data.area.mean()
std = house_data.area.std()
upper_limit = mean + std*3
lower_limit = mean - std*3
print("upper_limit : ", upper_limit)
print("lower_limit : ",lower_limit)

sns.boxplot(house_data["area"], ax=ax_box)
ax_box.axvline(mean, color='b', linestyle='--')


sns.distplot(house_data["area"], ax=ax_hist,kde=False,norm_hist=True)
ax_hist.axvline(lower_limit, color='r', linestyle='--')
ax_hist.axvline(mean, color='b', linestyle='-')
ax_hist.axvline(upper_limit, color='r', linestyle='--')



plt.legend({'Mean':mean})

ax_box.set(xlabel='')

x = np.linspace(mean - 3*std, mean + 3*std, 100)
plt.plot(x, norm.pdf(x, mean, std))
plt.legend({"lower_limit":lower_limit,'Mean':mean,'upper_limit':upper_limit})
print("Mean Value : ",mean)
print("Median Value : ",median)
print("1st Standard Deviation : ",std)

plt.show()
```

```python
area_outlier = house_data[(house_data.area>upper_limit) | (house_data.area<lower_limit)]
area_outlier
```

```python
house_data['area'].values[house_data['area'] > upper_limit] = house_data.area.mean()
house_data['area'].values[house_data['area'] < lower_limit] = house_data.area.mean()
```

```python
#Encoding categorical data - mainroad, guestroom, basement, hotwaterheating, airconditioning, furninshing status
house_data.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
encoder_mainraod = pd.DataFrame(encoder.fit_transform(house_data[['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning','furnishingstatus']]).toarray())
clean_house_data = house_data.join(encoder_mainraod)
clean_house_data.head()
```

```python
#dropping original columns
clean_house_data.drop('mainroad', axis=1, inplace=True)
clean_house_data.drop('guestroom', axis=1, inplace=True)
clean_house_data.drop('basement', axis=1, inplace=True)
clean_house_data.drop('hotwaterheating', axis=1, inplace=True)
clean_house_data.drop('airconditioning', axis=1, inplace=True)
clean_house_data.drop('furnishingstatus', axis=1, inplace=True)
clean_house_data.head()
```

```python
#Spliting the dataset into independent and dependent variables
x = clean_house_data.iloc[:,1:].values
y = clean_house_data.iloc[:,0].values
```

```python
#Scaling independent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled = sc.fit(x)
x_scaled = scaled.transform(x)
x
```

```python
#Spliting into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Building the model - Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error

model_LR = LinearRegression()
model_LR.fit(x_train, y_train)
Y_pred = model_LR.predict(x_test)

print(mean_absolute_percentage_error(y_test, Y_pred))
```

```python
#Building the model - SVR Regression

from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_percentage_error

model_SVR = svm.SVR()
model_SVR.fit(x_train,y_train)
Y_pred = model_SVR.predict(x_test)

print(mean_absolute_percentage_error(y_test, Y_pred))
```

```python
#Building the model - Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

model_RFR = RandomForestRegressor(n_estimators=10)
model_RFR.fit(x_train, y_train)
Y_pred = model_RFR.predict(x_test)

mean_absolute_percentage_error(y_test, Y_pred)
```
